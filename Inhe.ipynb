```python
class Animal():
    def __init__(self, eyes, hasMouth, teeth, gender):
        self.__eyes = eyes
        self.__hasMouth = hasMouth
        self.__gender = gender
        self.__teeth = teeth

    def getEyes(self):
        return self.__eyes
    def getHasMouth(self):
        return self.__hasMouth
    def getGender(self):
        return self.__gender
    def getTeeth(self):
        return self.__teeth
    def setEyes(self, eyes):
        self.__eyes = eyes
    def setHasMouth(self, hasMouth):
        self.__hasMouth = hasMouth
    def setGender(self, gender):
        self.__gender = gender
    def setTeeth(self, teeth):
        self.__teeth = teeth
    def talk(self):
        print("Talking")
    def hunt(self):
        print("Hunting")
    def eat(self):
        print("Eatting")
    def fight(self):
        print("Fighting")
```

```python
class Terrestrial(Animal):
    def __init__(self, eyes, hasMouth, teeth, gender, limbs):
        Animal.__init__(self, eyes, hasMouth, teeth, gender)
        self.__limbs = limbs
    def getLimbs(self):
        return self.__limbs
    def setLimbs(self, limbs):
        self.__limbs = limbs
    def run(self):
        print("Running")
    def walk(self):
        print("Walking")
```

```python
class Aquatic(Animal):
    def __init__(self, eyes, hasMouth, teeth, gender, fins, gills):
        super().__init__(eyes, hasMouth, teeth, gender)
        self.__fins = fins
        self.__gills = gills
    def getFins(self):
        return self.__fins
    def setFins(self, fins):
        self.__fins = fins
    def getGills(self):
        return self.__gills
    def setGills(self, gills):
        self.gills = gills
    def swim(self):
        print("Swimming")
```

```python
class Amphibian(Terrestrial, Aquatic):
    def __init__(self, eyes, hasMouth, teeth, gender, fins, gills, limbs):
        Terrestrial.__init__(self, eyes, hasMouth, teeth, gender, limbs)
        Aquatic.__init__(self, eyes, hasMouth, teeth, gender, fins, gills)
```

```python
class Cat(Terrestrial):
    def __init__(self, furColor, eyes, hasMouth, teeth, gender, limbs):
        super().__init__(eyes, hasMouth, teeth, gender, limbs)
        self.__furColor = furColor
    def getFurColor(self):
        return self.__furColor
    def setFurColor(self, furColor):
        self.__furColor = furColor
    def eat(self):
        print("Eatting rat")
    def talk(self):
        print("Meow Meow")
```

```python
class Dog(Terrestrial):
    def __init__(self, furColor, eyes, hasMouth, teeth, gender, limbs):
        super().__init__(eyes, hasMouth, teeth, gender, limbs)
        self.__furColor = furColor
    def getFurColor(self):
        return self.__furColor
    def setFurColor(self, furColor):
        self.__furColor = furColor
    def eat(self):
        print("Eatting rat")
    def talk(self):
        print("Bhaau Bhaau")
```

```python
class Shark(Aquatic):
    def __init__(self, color, eyes, hasMouth, teeth, gender, fins, gills, limbs):
        super().__init(eyes, hasMouth, teeth, gender, fins, gills, limbs)
        self.__color = color
    def getColor(self):
        return self.__color
    def setColor(self, color):
        self._color = color
    def eat(self):
        print("Eatting small fishes")
```

```python
class Frog(Amphibian):
    def __init__(self, eyes, hasMouth, teeth, gender, fins, gills, limbs, color):
        super().__init__(eyes, hasMouth, teeth, gender, fins, gills, limbs)
        self.__color = color
    def getColor(self):
        return self.__color
    def setColor(self, color):
        self.__color = color
    def eat(self):
        print("Eatting flies and insects")
```

```python
def simple():
    print('In Simple Inheritance, the derived class inherits from a base class. In our case, Terrestrial class inherits Animal class. By doing this, Terrestrial class inherits all the properties of Animal class.')
    t = Terrestrial(2, True, 10, "Female", 4)
    print("----Created a instance of Terrestrial class----")
    print("Characteristics of Animal class:-")
    print("No of eyes : ", t.getEyes())
    print("Does animal has mouth? =>", t.getHasMouth())
    print("No. of teeth : ", t.getTeeth())
    print("Gender - ", t.getGender())
    print("No. of limbs : ", t.getLimbs())
```

```python
def multiple():
    print("----In Multiple inheritance, derived class inherits from more than one base class----")
    print("Amphibian class inherits from Terrestrial and aquatic classes. This is an example of Multiple inheritance. It will have the properties of both the classes.")
    a = Amphibian(2, True, 15, "Male", 0, 2, 2)
    print("----Created a instance of Amphibian class----")
    print("Characteristics of Amphibian class:-")
    print("No of eyes : ", a.getEyes())
    print("Does animal has mouth? =>", a.getHasMouth())
    print("No. of teeth : ", a.getTeeth())
    print("Gender - ", a.getGender())
    print("No. of limbs : ", a.getLimbs())
    print('No. of gills : ', a.getGills())
    print("No. of fins : ", a.getFins())
```

```python
def multiLevel():
    print("Cat class has inherited from Terrestrial class which in turn has inherited from Animal class. This is a example of Multi level inheritance.")
    c = Cat("White and Black", 2, True, 20, "Female", 4)
    print("----Created a instance of Cat class----")
    print("Characteristics of Cat class:-")
    print("No of eyes : ", c.getEyes())
    print("Does animal has mouth? =>", c.getHasMouth())
    print("No. of teeth : ", c.getTeeth())
    print("Gender - ", c.getGender())
    print("No. of Limbs : ", c.getLimbs())
    print("Fur color : ", c.getFurColor())
```

```python
def herarchial():
    print("In Herarchial inheritance, there is one base class and more than one derived class")
    print("----Created instances of Terrestial and aquatic class----")
    t = Terrestrial(2, True, 10, "Female", 4)
    q = Aquatic(2, True, 7, "Male", 4, 2)
    print("---Characteristics of Terrestial class---")
    print("No of eyes : ", t.getEyes())
    print("Does animal has mouth? =>", t.getHasMouth())
    print("No. of teeth : ", t.getTeeth())
    print("Gender - ", t.getGender())
    print("No. of limbs : ", t.getLimbs())
    print("---Characteristics of Aquatic class---")
    print("No of eyes : ", q.getEyes())
    print("Does animal has mouth? =>", q.getHasMouth())
    print("No. of teeth : ", q.getTeeth())
    print("Gender - ", q.getGender())
    print("No. of gills : ", q.getGills())
    print("No. of fins : ", q.getFins())
```
