```python
class Bank():
    def __init__(self, nameOfBank:str):
        self.__name = nameOfBank
        self.__accounts = dict()
    def add_account(self, user:BankAccount):
        self.__accounts[user.getAccountNo()] = user
    def is_account_present(self, acc:int):
        if self.__accounts.get(acc) == None :
            return False
        else:
            return True
    def displayAccounts(self):
        print(self.__accounts)
    def getUser(self, acc:int):
        return self.__accounts.get(acc)
    def is_password_valid(self, acc:int, password:str):
        user = self.__accounts.get(acc)
        if(password == user.getPassword()):
            return True
        else:
            return False

class BankAccount():
    def __init__(self, userName:str, bal:float, password:str, bank:Bank):
        self.__userName = userName
        self.__balance = bal
        self.__password = password
        self.__accountNo = BankAccount.generateAccountNo(bank)
        bank.add_account(self)
    def deposit(self, money:float):
        self.__balance += money
    def withdraw(self, money:float):
        if(self.__balance < money):
            print("Withdrawal of Rs ", money," not possible due to insufficient balance.")
        else:
            self.__balance -= money
            print("Rs ", money, " withdraw completed.")
    def checkBalance(self):
        print("Available Balance - Rs ", self.__balance)
    def changePassword(self):
        oldPassword = ""    
        while oldPassword != self.__password:
            oldPassword = input("Enter your previous Password: ")
        self.__password = input("Enter your new Password: ")
        print("Password changed successfully")
    def getAccountNo(self):
        return self.__accountNo
    def getDescription(self):
        print('Name - ', self.__userName)
        print('Current Balance - Rs ', self.__balance)
        print('Account No. - ', self.__accountNo)
    def getPassword(self):
        return self.__password
    @classmethod
    def generateAccountNo(cls, bank:Bank):
        acc = random.randint(10 ** 7, 10 ** 8 - 1)
        while bank.is_account_present(acc):
            acc = random(8)
        return acc
            
        
        
```

```python
def take_new_user_info():
    print('Thank You for choosing COEP BANK')
    print("---Creating New Account---")
    name = input('Enter Your Name : ')
    balance = float(input('Initial Balance: '))
    password = input('Enter Your Password: ')
    return (name, balance, password)
```

```python
option = 0
b = Bank('COEP BANK')
while option != 3:
    print("Welcome to COEP BANK")
    print('1. Login')
    print('2. New User')
    print('3. Exit')
    option = int(input('Enter your option: '))
    if option == 1:
        accountNo = int(input('Enter your account No.: '))
        pin = input('Enter your password: ')
        if(b.is_account_present(accountNo)):
            if(b.is_password_valid(accountNo, pin)):
                user = b.getUser(accountNo)
                #show menu for functionalities
                while option != 6:
                    print('1. Check Balance')
                    print('2. Deposit')
                    print('3. Withdrawal')
                    print('4. Profile')
                    print('5. Change Password')
                    print('6. Logout')
                    option = int(input('Enter your option: '))
                    if(option == 1):
                        user.checkBalance()
                    elif option == 2:
                        money = float(input('Enter the amount of money to be deposited: Rs '))
                        user.deposit(money)
                    elif option == 3:
                        money = float(input('Enter the amount of money to be withdraw: Rs'))
                        user.withdraw(money)
                    elif(option == 4):
                        user.getDescription()
                    elif(option == 5):
                        user.changePassword()
                    elif option == 6:
                        print('---Logging Out---')
                        break
                    else:
                        print('You entered an invalid option.')
                        print('Please choose a valid option.')
                        
            else:
                print("Password is invalid.")
        else:
            print('Account No. is invalid.')
    elif option == 2:
        name, balance, pin = take_new_user_info()
        user = BankAccount(name, balance, pin, b)
        print('---Account Created Successfully---')
        print('Your account no. is ', user.getAccountNo())
        print('---Please Login to avail services---')
        user = None
    elif option != 3:
        print('!!You have entered an invalid option!!')
```
